### 4. PassengerId열을 통해 Group열과 Number열 만들기

```python
train_group = []
train_number = []
for i in range(len(train['PassengerId'])):
    g = train['PassengerId'].iloc[i].split('_')[0]
    n = train['PassengerId'].iloc[i].split('_')[1]
    train_group.append(g)
    train_number.append(n)

train['Group'] = train_group
train['Number'] = train_number

train = train.drop('PassengerId', axis = 'columns')
train
```

```python
test_group = []
test_number = []
for i in range(len(test['PassengerId'])):
    g = test['PassengerId'].iloc[i].split('_')[0]
    n = test['PassengerId'].iloc[i].split('_')[1]
    test_group.append(g)
    test_number.append(n)

test['Group'] = test_group
test['Number'] = test_number

test = test.drop('PassengerId', axis = 'columns')
test
```

### 6. Cabin열을 통해 deck, num, side 열 만들기

```python
deck_group = []
num_group = []
side_group = []

for i in range(len(train['Cabin'])):
    d = train['Cabin'][i].split('/')[0]
    n = train['Cabin'][i].split('/')[1]
    s = train['Cabin'][i].split('/')[2]
    deck_group.append(d)
    num_group.append(n)
    side_group.append(s)

train['deck'] = deck_group
train['num'] = num_group
train['side'] = side_group
               
train = train.drop('Cabin', axis = "columns")
train
```

```python
test['deck'] = test['Cabin'].str.split('/').str[0]
test['num'] = test['Cabin'].str.split('/').str[1]
test['side'] = test['Cabin'].str.split('/').str[2]

test = test.drop('Cabin', axis = "columns")
test
```

## 2. Modeling

### 1. library 호출

```python
import sklearn
import catboost
import optuna
```

### 2. seed 설정

```python
import random
seed = random.seed(17)
```

### 3. train_test_split

```python
X = train.drop('Transported', axis=1)
Y = train['Transported']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=seed)
```

### 4. Catboost pool함수를 이용해 train, validation 데이터를 pool객체로 만들기

```python
# pool은 여러 입력 값에 걸쳐 함수의 실행을 병렬 처리하고 입력 데이터를 프로세스에 분산시키는 편리한 방법을 제공합니다(데이터 병렬 처리)
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import accuracy_score

cat_cols = train.columns[train.dtypes == 'object'].to_list()

train_pool = Pool(x_train, y_train, cat_features = cat_cols)
val_pool = Pool(x_val, y_val, cat_features = cat_cols)
```

### 5. CatBoostClassifier 모델 만들어 훈련

```python
model = CatBoostClassifier(eval_metric = "Accuracy", early_stopping_rounds = 200, random_seed = seed)
model.fit(train_pool, eval_set=val_pool, plot=True)
```

### 6. 모델 성능을 알아보기 위한 validation set에서의 accuracy 계산

```python
model.best_score_
```

```python
accuracy = accuracy_score(y_val.astype(str), model.predict(x_val))
accuracy
```

### 7. Optuna library를 이용해 parameter 튜닝을 위한 objective함수 만들기, accuracy 반환

```python
def objective(trial):
    param = {"random_state" : 17,
            "learning_rate" : trial.suggest_loguniform('learning_rate', 0.01, 0.3),
            'bagging_temperature' : trial.suggest_loguniform('bagging_temperature', 0.01, 100.00),
            "n_estimators" : trial.suggest_int("n_estimators", 1000, 10000),
            "max_depth" : trial.suggest_int("max_depth", 0, 16),}
    
    opt_model = CatBoostClassifier(**param, eval_metric = "Accuracy", early_stopping_rounds = 200, random_seed = seed)
    opt_model.fit(train_pool, eval_set=val_pool, plot=True)
    
    opt_accuracy = accuracy_score(y_val.astype(str), opt_model.predict(x_val))
    return opt_accuracy
```

### 8. Optuna 결과 재현을 위해 TPESampler(seed=17) 사용한 파라미터 튜닝

```python
from optuna.samplers import TPESampler

sampler = TPESampler(seed=17)
study = optuna.create_study(
    direction="maximize",
    sampler=sampler,
)

study.optimize(objective, n_trials=5)
```

```python
best_param = study.best_trial.params
best_param
```

### 9.  튜닝한 parameter 이용하여 새로운 classifier 만들어 결과 비교

```python
best_model = CatBoostClassifier(**best_param, eval_metric = "Accuracy", early_stopping_rounds = 200, random_seed = seed)
best_model.fit(train_pool, eval_set=val_pool, plot=True)
```

```python
best_model.best_score_
```

```python
accuracy_score(y_val.astype(str), best_model.predict(x_val))
```
