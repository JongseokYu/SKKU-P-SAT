## Chaper1. API 사용

#### 문제1. 영화목록 API를 이용하여 감독의 이름이 '봉준호'인 영화목록으로 데이터 프레임 생성
#### 'prdtYear'변수를 기준으로 오름차순 정렬
#### directors와 companys 칼럼은 이름만 남김

```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=e2cf363075e3fc17078a3d35a01a0eee&directorNm=봉준호'
res = requests.get(url)
text = res.text

data = json.loads(text)
data = pd.DataFrame(data['movieListResult']['movieList'])
data = data.sort_values('prdtYear', ascending = True)
data = data.reset_index(drop = True)

### apply lambda를 통한 형태 변화
data['directors'] = data['directors'].apply(lambda x : [tmp['peopleNm'] for tmp in x])
data['companys'] = data['companys'].apply(lambda x : [tmp['companyNm'] for tmp in x])

data
```

```python
### url 받아오기
url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=e2cf363075e3fc17078a3d35a01a0eee&directorNm=봉준호'
res = requests.get(url)

### text로 변환
text = res.text

### data로 저장
json_data = json.loads(text)
data = pd.DataFrame(json_data['movieListResult']['movieList'])
data = data.sort_values('prdtYear', ascending = True)
data = data.reset_index(drop = True)

### directors 열
data['directors'] = data['directors'].apply(lambda x : str(x).replace("[{'peopleNm': '봉준호'}]",'[봉준호]'))

### companys 열
firm = []
for i in range(10) :
    if json_data['movieListResult']['movieList'][i]['companys']  == [] :
        firm.append([])
    else :
        firm1 = '[' + json_data['movieListResult']['movieList'][i]['companys'][0]['companyNm'] +']'
        firm.append(firm1)
        
data['companys'] = firm
data
```

#### 문제2. 일별 박스오피스 API를 이용하여 2022년 전반기의 일별 박스오피스 데이터 생성
#### 데이터 csv로 저장

```python
# 빈 데이터프레임 만들기
result = pd.DataFrame()

# date range 설정
date_range = pd.date_range(start = "20220101", end = "20220630")

# date -> str형태로 바꾸고 list로 바꿔주기
date = date_range.strftime("%Y%m%d").tolist()

# data 받기
for i in range(len(date)):
    url = "http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=e2cf363075e3fc17078a3d35a01a0eee&targetDt={}".format(date[i])
    res = requests.get(url)
    json_data = json.loads(res.text)
    data = pd.DataFrame(json_data['boxOfficeResult']['dailyBoxOfficeList'])
    showrange = json_data['boxOfficeResult']['showRange']
    data['showRange'] = showrange
    result = pd.concat([result, data])

result
```

### 2021년 전반기 월간 차트

```python
### url 가져오기
url = "http://www.melon.com/chart/index.htm"
driver.get(url)
```

```python
import time

driver.find_element_by_xpath('//*[@id="gnb_menu"]/ul[1]/li[1]/div/ul/li[4]/a/span').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="conts"]/div[3]/div/button').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="conts"]/div[3]/div/div/dl/dt/button[1]/span/span').click()
time.sleep(1)

# 각 요소별 list 만들기
title_list = []
singer_list = []
album_list = []
diff_list = []
rank_list = []
like_list = []

for m in range(1,6+1):
    driver.find_element_by_xpath('//*[@id="conts"]/div[3]/div/div/dl/dd[1]/ul/li[{}]/a/span'.format(m)).click()
    time.sleep(1)
    
    chart = driver.page_source
    soup = BeautifulSoup(chart, 'html.parser')
                                 
    titles = soup.find_all('div', attrs={'class': 'ellipsis rank01'})
    singers = soup.find_all('div', attrs={'class': 'ellipsis rank02'})
    albums = soup.find_all('div', attrs={'class': 'ellipsis rank03'})
    diffs = soup.find_all('span', attrs={'class': 'rank_wrap'})
    ranks = soup.find_all('span', attrs={'class': 'rank'})
    likes = soup.find_all('button', attrs={'class': 'button_etc like'})
    
    for title in titles :
        title_list.append(title.get_text(strip=True))
        
    for singer in singers :
        singer_list.append(singer.get_text(strip=True))
        
    for album in albums : 
        album_list.append(album.get_text(strip=True))
    
    for diff in diffs :
        diff_list.append(diff.get_text(strip = True))
        
    for rank in range(1,101) : 
        rank_list.append(rank)
        
    for like in likes : 
        l = like.get_text(strip = True)
        l = l.replace('좋아요총건수', '')
        like_list.append(l)
    
    dic = {'year' : 2021,
          'month' : m,
          'rank' : rank_list,
          'diff' : diff_list,
          'title' : title_list,
          'singer' : singer_list,
          'album' : album_list,
          'like' : like_list}
    
    data = pd.DataFrame(dic)
    result = data

result
```
